```python
import pandas as pd
#import time
#import random

snpid_df=pd.read_csv('input_data.csv')
snpid_list=snpid_df['snpid'].tolist()
dataframe_list=[]
failed_snpid=[]
total_number_of_snpid=len(snpid_list)
for snpid in snpid_list[50000:56125]:
    try:  
        url = 'https://www.snpedia.com/api.php?action=parse&page='+snpid+'&format=json'
        dfs = pd.read_html(url)
        
        #Adding some random waiting time so that server doesnt gets loaded
        #time.sleep(random.randint(0,7))
        
        dictionary_for_df={}
        dictionary_for_df['snpid'] = snpid
        for df in dfs:
            if df[0][0] == ('Position' or 'position'):
                dictionary_for_df['position']=df[1].tolist()[0]
            if df[0][0] == ('Chromosome' or 'chromosome'):
                dictionary_for_df['chromosome']=df[1].tolist()[0]  
            
            if dfs[0].shape == (1,1):
                dictionary_for_df['final_summary']=dfs[0][0][0]  
            else:
                dictionary_for_df['final_summary']='Nan' 

                try:
                    df[0][0].split("\\")[0] == ('Geno' or 'geno')
                except Exception:
                    df[0][0].split("\\")[0] == 'Nan'
                    
            if df[0][0].split("\\")[0] == ('Geno' or 'geno'):
                try:
                    dictionary_for_df['geno1']=df[[0,2]].to_dict()[0][1].split("\\")[0]
                except Exception:
                    dictionary_for_df['geno1']='Nan'
                try:
                    dictionary_for_df['geno2']=df[[0,2]].to_dict()[0][2].split("\\")[0]
                except Exception:
                    dictionary_for_df['geno2']='Nan'
                try:
                    dictionary_for_df['geno3']=df[[0,2]].to_dict()[0][3].split("\\")[0]
                except Exception:
                    dictionary_for_df['geno3']='Nan'
    
                
                
                try:
                    dictionary_for_df['sum_1']=df[[0,2]].to_dict()[2][1].split("\\")[0]
                except Exception:
                    dictionary_for_df['sum_1']='Nan'
                try:
                    dictionary_for_df['sum_2']=df[[0,2]].to_dict()[2][2].split("\\")[0]
                except Exception:
                    dictionary_for_df['sum_2']='Nan'
                try:
                    dictionary_for_df['sum_3']=df[[0,2]].to_dict()[2][3].split("\\")[0]
                except Exception:
                    dictionary_for_df['sum_3']='Nan'
                
        dataframe_list.append(dictionary_for_df)
    
        print('Done ============> {} out of {} given entries'.format(snpid_list.index(snpid)+1,total_number_of_snpid))
        
    except Exception:
        failed_snpid.append(snpid)
        print("************** SNPID FAILED : {} ****************".format(snpid))

pd.DataFrame(dataframe_list)
```

```python
pd.DataFrame(dataframe_list).to_csv("output.csv")
```
